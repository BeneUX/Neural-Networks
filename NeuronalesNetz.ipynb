```python
import pandas as pd
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense, Input
```

```python
df = pd.read_csv('statics/housepricedata.csv')
df
```

```python
# Was sind Input was sind Output daten, unterteilen in x und y
# iloc schneidet dataframe, die ersten 10 Spalten in x, spalte 11 in y

x = df.iloc[:, 0:10]
y = df.iloc[:, 10]
```

```python
# Daten transformieren, beziehungsweise skalieren zwischen 0 und 1 mithilfe von sklearn

x = preprocessing.MinMaxScaler().fit_transform(x)
```

```python
# Datensatz auf Testdaten und Trainingsdaten aufsplitten, mithilfe von sklearn
#train_test_split(input wert, input wert, testsize in Prozent, weitere Möglichkeit anzugeben welche Daten genommen werden sollen)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
# Modell definieren mithilfe von keras

model = Sequential()

# Layer hinzufügen, Input Layer mit der expliziten Angabe der ankommenden Spalten
model.add(Input(shape=(10, )))
# Layer hinzufügen, Dense Layer(Anzahl Units, Aktivierungsfunktion)
# Single Layer könnte nur linear separierbare Probleme lösen, für ein XOR bräuchte man schon mindestens eine Hidden Layer
model.add(Dense(32, activation='relu'))
model.add(Dense(32, activation='relu'))
# Output Layer
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='mse', metrics=['accuracy'])
model.fit(x_train, y_train, batch_size=32, epochs=10, validation_data=(x_test, y_test))
```

```python
# Testen des Modells

test = [[8450, 7, 5, 856, 2, 1, 3, 8, 0, 548], # 1
        [6120, 7, 8, 832, 1, 0, 2, 5, 0, 576]] # 0

test = preprocessing.MinMaxScaler().fit_transform(test)

# Ausgabe des tests
print(model.predict(test))

# Speichern des Modells 
model.save("ann.h5")
```
